```python
import pyarrow
import pandas as pd
import os, re, time
from pyfitparquet import transformer
```

```python
# USER: must set DATA_DIR to location of .fit/.tcx files
# Defaults PARQUET_DIR to subdirectory within DATA_DIR

DATA_DIR = '../../data'
PARQUET_DIR = DATA_DIR + '/parquet'

try: os.mkdir(PARQUET_DIR) 
except FileExistsError: pass
```

```python
# Serializes all files in DATA_DIR to parquet
pyfitparq = transformer.PyFitParquet()
pyfitparq.data_to_parquet(DATA_DIR)
```

```python
# Opens transformed parquet files into
# dataframes and display their summaries

dframes = {}
for file in os.listdir(PARQUET_DIR):
#{
    matchobj = re.match('(\w+).parquet', file)
    if matchobj:
        pfile = f'{PARQUET_DIR}/{file}'
        print(pfile + ":")
        pdframe = pd.read_parquet(pfile, engine='pyarrow')
        dframes[pfile] = pdframe
        display(pdframe)
        #display(pdframe[60:120])
        print()
#}
```

```python
# Test re-initialization from configuation file
# USER: make some change to parquet_config.yml

pyfitparq.reset_from_config()
pyfitparq.data_to_parquet(DATA_DIR)
```
